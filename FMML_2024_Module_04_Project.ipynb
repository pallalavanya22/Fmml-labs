### Slope & Derivatives

Creating a function for the derivative of $f(x)$ called df(x)

````python
def df(x):
    return 2 * x + 1
````

### Gradient Descent

````python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell

new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001
x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(new_x)  # [REQUIRED]
    new_x = new_x - step_multiplier * gradient  # [REQUIRED]
    step_size = abs(new_x - previous_x)
    # print(step_size)
    x_list.append(new_x)
    slope_list.append(df(new_x))  # [REQUIRED]

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
````

## Gradient Descent

````python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
````

````python
## TASK-7 : Plot reduction in cost for each iteration
import matplotlib.pyplot as plt
import numpy as np

# Your existing functions g(x) and dg(x) remain unchanged

def gradient_descent(derivative_func, initial_guess,
                    multiplier=0.01, precision=0.0001, max_iter=300):
    """
    Calculates the local minimum of a function using gradient descent.

    Args:
        derivative_func (function): The derivative of the function to minimize.
        initial_guess (float): The starting point for the algorithm.
        multiplier (float, optional): The learning rate or step size. Defaults to 0.01.
        precision (float, optional): The desired precision for the minimum. Defaults to 0.0001.
        max_iter (int, optional): The maximum number of iterations. Defaults to 300.

    Returns:
        tuple: A tuple containing the local minimum, the list of x values,
               the list of derivative values, and the list of costs at each iteration.
    """

    x_list = [initial_guess]
    deriv_list = [derivative_func(initial_guess)]
    costs = [g(initial_guess)]  # Initialize costs with the initial cost

    for n in range(max_iter):
        previous_x = x_list[-1]
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient
        step_size = abs(new_x - previous_x)

        x_list.append(new_x)
        deriv_list.append(derivative_func(new_x))
        costs.append(g(new_x))  # Calculate and store the cost at each iteration
````

````python
## TASK-8 : Code the same things as they were done for the first two examples

import matplotlib.pyplot as plt
import numpy as np

# Make some data
x_2 = np.linspace(-2, 2, 1000)

# Define the function and its derivative
def g(x):
    return x**2 + 2*x + 1

def dg(x):
    return 2*x + 2

# Create a gradient descent object
class GradientDescent:
    def __init__(self, learning_rate=0.1, precision=0.00001):
        self.learning_rate = learning_rate
        self.precision = precision

    def run(self, x0):
        x_list = [x0]
        slope_list = [dg(x0)]
        cost_list = [g(x0)]

        for n in range(500):
            previous_x = x_list[-1]
            gradient = dg(previous_x)
            new_x = previous_x - self.learning_rate * gradient
            step_size = abs(new_x - previous_x)

            x_list.append(new_x)
            slope_list.append(dg(new_x))
            cost_list.append(g(new_x))

            if step_size < self.precision:
                print('Loop ran this many times:', n)
                break

        return x_list, slope_list, cost_list

# Run the gradient descent
gd = GradientDescent()
x0 = 3
x_list, slope_list, cost_list = gd.run(x0)

# Print the results
print('Local minimum occurs at:', x_list[-1])
print('Slope or dg(x) value at this point is:', slope_list[-1])
print('g(x) value or cost at this point is:', cost_list[-1])

# Plot the function and its derivative
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_2, g(x_2))
plt.plot(x_list, [g(x) for x in x_list], 'ro-')
plt.title('Function g(x)')
plt.xlabel('x')
plt.ylabel('g(x)')

plt.subplot(1, 2, 2)
plt.plot(x_2, dg(x_2))
plt.plot(x_list, slope_list, 'ro-')
plt.title("Derivative dg(x)")
plt.xlabel('x')
plt.ylabel("dg(x)")

plt.tight_layout()
plt.show()

# Plot the cost at each iteration
plt.figure(figsize=(10, 6))
plt.plot(cost_list)
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Reduction in Cost for Each Iteration')
plt.show()
````

#### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
#### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
#### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

#### $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

#### $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

#### MSE & Gradient Descent

````python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
````
